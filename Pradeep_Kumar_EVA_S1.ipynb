```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

# Lower Horizintal edge detector

This should have upper row and lower row set to extreme values

```
kernel = np.float32([[-1,-1,-1], [0,0,0], [1,1,1]])
LowerhorizontalEdges = cv2.filter2D(image, -1, kernel)
cv2_imshow(LowerhorizontalEdges)
```

# Upper Horizintal edge detector

This should have upper row and lower row set to extreme values

```

kernel = np.float32([[1,1,1], [0,0,0], [-1,-1,-1]])
UpperhorizontalEdges = cv2.filter2D(image, -1, kernel)
cv2_imshow(UpperhorizontalEdges)
```

```
kernel = np.float32([[0,1,1], [-1,0,1], [-1,-1,0]])
DegreeAngleDetector = cv2.filter2D(image, -1, kernel)
cv2_imshow(DegreeAngleDetector)
```

# 45 Degress Angle detector
To detect an 45 degree edge 

Kernel should be able to identfy the change of value from positive to zero and then to negative.

  0   1   1
 -1   0   1
 -1  -1   0

```
kernel = np.float32([[0,-1,-1], [1,0,-1], [1,1,0]])
DegreeAngleDetector = cv2.filter2D(image, -1, kernel)
cv2_imshow(DegreeAngleDetector)
```

# Blurring

A blurr image can be created by having the resultant image having each pixel as the average value of all its neighbouring pixels

    1 1 1
